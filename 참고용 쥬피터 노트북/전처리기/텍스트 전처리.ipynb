```python
import pandas as pd
```

# 파일 불러오기

```python
data = pd.read_csv('최종학습데이터.csv')
data = data[['desc', 'label']]
data.dropna(how='any', inplace=True)
data
```

# 특수문자 제거 및 개행문자 정리 함수

```python
import re

def 문장_전처리(문장):
    문장 = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$&\\\=\(\'\"\♥\♡\※\■\▼\§\★\●\⊙\＆\＠\☆\○\◎\◇\◆\■\△\▲\▽\→\←\↑\↓\↔\◁\◀\▷\▶\♤\♠\♧\♣\◈\▣\☎]','',str(문장))
    문장 = re.sub('\n*\n', '\n', 문장)
    return 문장
```

# 이모티콘 제거 함수

```python
def deEmojify(inputString):
    return inputString.encode('euc-kr', 'ignore').decode('euc-kr')
```

```python
data['desc'] = data['desc'].apply(문장_전처리)
```

```python
data['desc'] = data['desc'].apply(deEmojify)
```

```python
for i in data['desc']:
    print(i)
    print('\n')
```

```python
data.to_csv('최종학습데이터_전처리완.csv')
```
