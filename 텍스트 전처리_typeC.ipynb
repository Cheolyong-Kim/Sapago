```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('번개맞은당근_갤럭시s8.csv')
data['label'] = 1
data = data[['desc', 'label']]
```

```python
import re

re_list = []
for i in data['desc']:
    re_list.append(re.sub('\n*\n', '\n', i))
data['RE'] = re_list
```

```python
data['tokenized_sentence'] = data['RE'].apply(lambda x: x.split('\n'))
```

```python
status_w_l = ['기스', '배터리', '깨끗', '새', '미개봉', '새제품', '새상품', '모서리', '거의', '액정', '흠집', '얼룩', '충전', '하자', '멍', '콕', 'S급', 'SS급', '리퍼', 'A급', 'B급', '긁힘', '테두리', '개봉' ,'파손', '잔상', '뒷판']
```

```python
def status_detector(token_list):
    status_sentence_list = []
    for sentence in token_list:
        for word in status_w_l:
            if sentence.find(word) >= 0:
                status_sentence_list.append(sentence)
    output = list(set(status_sentence_list))
    return output
```

```python
data['status_sentence'] = data['tokenized_sentence'].apply(status_detector)
```

```python
def to_nan(sentence):
    if len(sentence) < 1:
        return np.nan
    else:
        return sentence
```

```python
data['status_sentence'] = data['status_sentence'].apply(to_nan)
```

```python
data = data.dropna(how='any')
```

```python
copy_data = data.copy()
copy_data['status_sentence'] = data['status_sentence'].apply(lambda x: ' '.join(x))
```

```python
def deEmojify(inputString):
    return inputString.encode('euc-kr', 'ignore').decode('euc-kr')
```

```python
copy_data['status_sentence'] = copy_data['status_sentence'].apply(deEmojify)
```

```python
new_list = []
for sentence in copy_data['status_sentence']:
    new_list.append(re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"\♥\♡\※\■\▼\§\★\●\⊙\＆\＠\☆\○\◎\◇\◆\■\△\▲\▽\→\←\↑\↓\↔\◁\◀\▷\▶\♤\♠\♧\♣]', '', sentence))
copy_data['status_sentence'] = new_list
```
